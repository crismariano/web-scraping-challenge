```python
# Import dependencies
import os
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
import requests
```

# NOTE: This notebook is for testing only and not for review.

```python
# URL of page to be scraped
url = 'http://mars.nasa.gov/news'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object
soup = bs(response.text, 'lxml')
type(soup)
```

```python
# Examine the results
# print(soup.prettify())
```

```python
# EXAMPLE OF NEWS TITLE

# <div class='content_title'>
# <a href='/news/8716/nasa-to-broadcast-mars-2020-perseverance-launch-prelaunch-activities/'>
# NASA to Broadcast Mars 2020 Perseverance Launch, Prelaunch Activities
# </a>
# </div>
```

```python
# Scrape NASA Mars News and collect the latest News Title and Paragraph Text.
# Assign the text to variables that you can reference later.

news_titles = soup.find_all('div', class_='content_title')

nasa_titles = []

for title in news_titles:
    nasa_titles.append(title.text.strip())
print(nasa_titles)
```

```python
title = soup.title.text.strip()
print(title)
```

```python
para = soup.find('div', class_='article_teaser_body')
print(para)
```

```python
# EXAMPLE OF PARAGRAPH

# <div class='rollover_description_inner'>
# Starting July 27, news activities will cover everything from mission engineering and science to returning samples from Mars to, of course, the launch itself.
# </div>
```

```python
# Paragraph Text

paragraphs = soup.find_all('div', class_='rollover_description_inner')
nasa_paragraphs = []
for paragraph in paragraphs:
    nasa_paragraphs.append(paragraph.text.strip())
print(nasa_paragraphs)
```

```python
# EXAMPLE OF IMAGES - https://www.jpl.nasa.gov/

# <footer>
#  <a class='button fancybox' data-description="This scene from NASA&#039;s Mars Exploration Rover Opportunity looks upward at &#039;Knudsen Ridge&#039; on the southern edge of &#039;Marathon Valley&#039; from inside the valley." 
#   data-fancybox-group='images' data-fancybox-href='/spaceimages/images/mediumsize/PIA20318_ip.jpg' data-link='/spaceimages/details.php?id=PIA20318' data-title='Steep &#039;Knudsen Ridge&#039; Along &#039;Marathon Valley&#039; on Mars' id='full_image'>
#   FULL IMAGE
#  </a>
# </footer>
```

```python
# Splinter
# JPL Mars Space Images - Featured Image

executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

image_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(image_url)


# HTML Object
html = browser.html
soup = bs(html, 'lxml')

images = soup.find('footer')
link = images.find('a')['data-fancybox-href']
featured_image_url = image_url + link

featured_image_url
```

```python
# Mars Hemispheres
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

hemi_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'

browser.visit(hemi_url)

hemi_html = browser.html
hemi_soup = bs(hemi_html, 'lxml')
```
